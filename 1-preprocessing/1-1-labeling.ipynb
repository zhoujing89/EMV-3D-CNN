根据更新的418套CT病理标注，划分为良性、浸润前、grade1、2、3、其他恶性，并存csv。

```python
import pandas as pd
# 读入结节信息
dat = pd.read_csv("/course75/RealData/结节（病理全）-给学生-0718.csv",encoding = 'GBK')
dat
```

```python
import numpy as np
dat['flag'] = None #添加一列表示良恶性
l0 = ['错构瘤','放线菌','肺内淋巴结','肺脓肿','肺泡蛋白沉积症','孤立性纤维瘤','良性','良性，肉芽肿','肉芽肿',
      '炎性增生','炎症','隐球菌','细支气管腺瘤','腺瘤'] # 良性
l1 = ['AAH','AIS','MIA'] # 腺癌浸润前病变
l2 = ['不典型类癌','典型类癌','大细胞神经内分泌','恶性','恶性，转移可能？','非小细胞肺癌','结肠腺癌肺转移',
      '类癌','鳞癌','神经内分泌癌','腺癌','腺癌（粘液腺癌）','腺鳞癌','小细胞','小细胞癌','粘液腺癌',
      '转移','转移性乳腺癌','转移性尿路系统癌','结肠腺癌肺转移'] # 恶性

for i in range(dat.shape[0]):
    if dat['Pathology'][i] in l0:
        dat['flag'][i] = '良性'
    elif dat['Pathology'][i] in l1:
        dat['flag'][i] = '浸润前病变'
    elif dat['Pathology'][i] in l2:
        if np.isnan(dat['病理分级'][i]):
            dat['flag'][i] = '其他恶性癌'
        elif dat['病理分级'][i] == 1.0:
            dat['flag'][i] = 'grade1'
        elif dat['病理分级'][i] == 2.0:
            dat['flag'][i] = 'grade2'
        elif dat['病理分级'][i] == 3.0:
            dat['flag'][i] = 'grade3'
```

```python
dat['flag'].value_counts()
```

```python
datflag = dat.dropna(axis=0,subset=['flag'], inplace=False).reset_index()[['ID', 'X', 'Y', 'Z','Size.cm.','flag',
                                                                           'lung','position','Image.type','Pathology']]
datflag # 最终有1075个标注数据
```
